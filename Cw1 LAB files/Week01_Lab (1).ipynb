### Typical 1. Total Time to Watch a Series

Compute total viewing time (no breaks) for a series:
- 5 seasons, 10 episodes per season
- Season 1: 45 minutes 35 seconds per episode
- Seasons 2–5: 35 minutes 15 seconds per episode
Display the total as hours, minutes, and seconds.

**Example Output:**
```
Total time: 39h 30m 0s
```


````python
# Your code here
s1= 10

s1_per_episode=45*60+35 # CONVERTING MINUTES INTO SEC

total_season1= s1*s1_per_episode # Finding how long season 1 was in seconds

ep_s2_5= 40
s2_s5_per_episode = 35*60 +15# CONVERTING MINUTES INTO SEC
total_s2_5 = ep_s2_5*s2_s5_per_episode# Finding how long season 2 was in seconds

total_seconds= total_season1+total_s2_5 # calculating the total seconds for s1 to s5

hours= total_seconds//3600 # Calculate the number of full hours in the total seconds

remainder=total_seconds//3600 # Calculate the remaining seconds after extracting hours

minutes=remainder//60 # Calculate the number of full minutes from the remainder

seconds= remainder%60 # The leftover seconds after hours and minutes are extracted

# Printing the result for hrs, minute and second
print(f'total time in the series from s1 to s5 is: {hours}hrs {minutes} min{seconds} sec')

````

### Typical 2. Savings Projection (3 Years)

Compute final savings after 3 years given:
- Save £10 every week for 3 years
- Year 1 bonus: +£20 each week
- Every 2 weeks: −£0.50 (sister)
Assume 52 weeks per year; use arithmetic and augmented assignment; no conditionals.

**Example Output:**
```
Final savings after 3 years: £4689.50
```


````python
# Your code here
saving_per_week=10
yr_1_bonus=20
sister=-0.50
weeks_in_a_year=52

first_year_savings=(saving_per_week+yr_1_bonus)*(weeks_in_a_year)+(sister*weeks_in_a_year/2)#calculating the first year saving  plus bonus gained for first year
second_year_savings=(saving_per_week*weeks_in_a_year)+(sister*weeks_in_a_year/2)#calculating the second year saving  
third_year_savings=(saving_per_week*weeks_in_a_year)+(sister*weeks_in_a_year/2)#calculating the third year saving  
total_saving_after_3yrs= first_year_savings + second_year_savings + third_year_savings # Adding the whole to find how much saving was there for 3years

# Printing the result of the first year , second year , third year and finally find the sum of all the year together 
print ('First year savings', first_year_savings) 
print ('Second year savings', second_year_savings) 
print('Third year savings', third_year_savings )
print(f'Final savings after 3 years:{total_saving_after_3yrs:.2f}')
````
