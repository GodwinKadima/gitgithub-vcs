4. **Body mass index (BMI)** is a measure of health and weight.

It can be calculated by taking your weight in kilograms and dividing it by the square of your height in meters. The interpretation of BMI for people 20 years or older is as follows:


BMI	Interpretation
BMI < 18.5	Underweight
18.5 ≤ BMI ≤ 25.0	Normal
25.0 ≤ BMI ≤ 30.0	Overweight
30.0 ≤ BMI 	Obese
 
Write a program that prompts the user to enter weight in pounds and height in inches and display the BMI. Note that one pound is 0.45359237 kilograms, and one inch is 0.0254 meters


````python
user_weight = float(input('Enter weight in pounds:'))# Asking the user to enter their weight in 
user_height = float(input('Enter height in inches:'))
kg = user_weight * 0.45359237 # Converting the weight in pounds entered by the user to kilograms
h = user_height * 0.0254 # Converting the height in inches entered by the user to meters
bmi = kg / h**2 # Calculating the BMI USING THE KILOGRAMS AND THE USER HEIGHT IS METERS

print('Your BMI Is',bmi)# printing the BMI calculate

# Defining the USER BMI WEIGHT
if bmi < 18.5 :
    print('UnderWeight')
elif 18.5 <= bmi <= 25.0 :
    print('Normal')
elif 25.0 <= bmi <= 30.0:
    print('Overweight')
elif bmi <= 30.0:
    print('Obese')
````

5. **Student Marks Analysis**
You are working as a data analyst in a school. You have the exam marks of students stored in a list. Your task is to analyze the data.

**Requirements**:

- Store marks of students in a list.

- Use a loop to calculate the average mark.

- Find the highest and lowest mark using list functions.

- Print the number of students who passed (marks ≥ 50) and failed.

````python
# Marks of students stored in a list
marks = [80, 20, 100, 49, 50, 12, 45]

# Calculate the Average Mark
total = 0
for mark in marks:
    total += mark  # Add each mark to the total

average = total / len(marks)  # Divide the total by the number of marks to get the average
print("Average mark:", average)

# Find the Highest and Lowest Mark
highest_mark = max(marks)  # Find the highest mark using the max() function
lowest_mark = min(marks)   # Find the lowest mark using the min() function

print("Highest mark:", highest_mark)
print("Lowest mark:", lowest_mark)

# Count the Number of Students Who Passed and Failed
passed_count = 0
failed_count = 0

for mark in marks:
    if mark >= 50:
        passed_count += 1  # Increase the passed_count for marks 50 or above
    else:
        failed_count += 1  # Increase the failed_count for marks below 50

print("Number of students who passed:", passed_count)
print("Number of students who failed:", failed_count)
````

12. **Dog Years**

It is commonly said that one human year is equivalent to 7 dog years. However, this simple conversion fails to recognise that dogs reach adulthood in approximately two years. As result, some people believe that it is better to count each of the first two human years as 10.5 dog years, and then count each additional human year as 4 dog years. 

Write a program that implements the conversion from human years to dog years described in the previous paragraph. Ensure that your program works correctly for conversions of less than two or more humans’ years. Your program should display an appropriate error message if the user enters a negative number. 


````python
# Ask the user for human years
human_years = float(input("Enter the dog's age in human years: "))

# Check for invalid input
if human_years < 0:
    print("Error: Age cannot be negative.")

# For the first two years
elif human_years <= 2:
    dog_years = human_years * 10.5
    print(f"The dog's age in dog years is {dog_years} years.")

# For more than two years
else:
    dog_years = (2 * 10.5) + (human_years - 2) * 4
    print(f"The dog's age in dog years is {dog_years} years.")
````

17. **Collecting Positive Numbers**
You are writing a small program to help a teacher keep track of positive numbers entered by students. The teacher also wants to know if any even numbers were given.

**Requirements:**

1. Keep asking the user to enter a positive number, or type `"done"` to stop.
2. Store all entered numbers in a **list**.
3. After the input ends:

   * If numbers were entered:

     * Print all the numbers entered.
     * Check the list and print the **first even number**. If no even number exists, print `"No even numbers found"`.
   * If no numbers were entered, print `"No numbers were entered"`.


> ### Note 
>
>* **`lower()`**: Converts any text into lowercase.
>  Example: `"DONE".lower()` becomes `"done"`.
>  This makes it easier to compare user input without worrying about uppercase/lowercase.
>
>* **`isdigit()`**: Checks if a string only contains digits (0–9).
>  Example: `"123".isdigit()` returns `True`, but `"12a".isdigit()` returns `False`.
>  This helps us make sure the user typed a valid number before converting it into an integer.
> 


---

````python
numbers = []

while True:
    user_input = input("Enter a positive number (or 'done' to stop): ")

    if user_input.lower() == 'done':
        break

    try:
        num = int(user_input)
        if num > 0:
            numbers.append(num)
        else:
            print("Please enter a positive number.")
    except ValueError:
        print("Invalid input. Please enter a number or 'done'.")

if not numbers:
    print("No numbers were entered")
else:
    print("Numbers entered:", numbers)

    first_even = None
    for num in numbers:
        if num % 2 == 0:
            first_even = num
            break

    if first_even is not None:
        print("First even number:", first_even)
    else:
        print("No even numbers found")
````

    1. If the user’s input matches the lottery in the exact order, the award is £10,000.
    2. If all the digits in the user’s input match all the digits in the lottery number, the award is £3,000.
    3. If one digit in the user’s input matches a digit in the lottery number, the award is £1,000.

20. (Find the two highest scores) - Write a program that prompts the user to enter the number of students and each student’s score and displays the highest and second highest scores.


````python
num_students = int(input("Enter the number of students: "))

scores = []
for i in range(num_students):
    score = float(input(f"Enter score for student {i+1}: "))
    scores.append(score)

if num_students < 2:
    print("Need at least two students to find the highest and second highest scores.")
else:
    highest = max(scores)
    scores.remove(highest)
    second_highest = max(scores)

    print("Highest score:", highest)
    print("Second highest score:", second_highest)
````
