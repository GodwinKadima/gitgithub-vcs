3. Create a function taking one parameters for a password. The function would convert each character of the password into corresponding ASCII numbers and return one string composed of the ASCII numbers instead

````python
# define function
def password(pw):
   result = ""
   for ch in pw:
      result += str(ord(ch)) # ord convert from character to ASCII value
   return result 

# Final Print converted character to  ASCII value 
print(password('Bye'))
````

10. Create a function (s) counting all lower case, upper case, digits, and special symbols from a given string and print them out.
   
```python

   example_password = "P@#yn26at^&i5ve"
```

````python
#define function
def count_characters(s):
    # The count start from 0 until user insert password
    lower = 0
    upper = 0
    digit = 0
    special = 0

    for char in s:
        if char.islower(): # Check if text contain lowercase
            lower += 1
        elif char.isupper(): # chaek if text contain  uppercase
            upper += 1
        elif char.isdigit(): # check if text contains number
            digit += 1
        else:
            special += 1 # check if text contains special character

    # print the result if true
    print("Lowercase letters:", lower)
    print("Uppercase letters:", upper)
    print("Digits:", digit)
    print("Special symbols:", special)


# Example run
text = "P@#yn26at^&i5ve"
count_characters(text)
````

16. (Binary to hex) Write a function that parses a binary number and converts itinto a hex number. The function header is: def binaryToHex(binaryValue):
Write a test program that prompts the user to enter a binary number and displays the corresponding hexadecimal value.



````python
def binaryToHex(binaryValue):

    # Converting to an integer base 2 to decimal base 10
    deci_num = int(binaryValue,2)

    #convert the decimal Value into hexadeciaml value
    hex_num = hex(deci_num)

    return hex_num # This will give out the output given by the hex_num

# Print convertion
binaryValue = input("Enter a binary:")
print("The Hex value is", binaryToHex(binaryValue)) # Call the function then print the return hexadecimal
````
