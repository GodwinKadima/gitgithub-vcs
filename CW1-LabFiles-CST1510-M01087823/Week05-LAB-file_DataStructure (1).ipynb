### Activity 9: Finding "Dream" Sentences in Les_Brown.txt

From the `Les_Brown.txt` file, read the content and identify all sentences that contain the word "dream". Write these sentences to a new file called `dream_sentences.txt`.

**Steps:**
1. Open and read the content of `Les_Brown.txt`.
2. Split the text into sentences.
3. Filter out sentences that contain the word "dream" (case-insensitive).
4. Write these sentences into the new file `dream_sentences.txt`.
<hr>

`````python
# Reading the file 
with open("Les_Brown.txt", "r") as f:
     read = f.readlines()

print('Reading file:\n',read,'\n') 

# Turning Text into Sentences
text = " ".join(read)
sentences = text.split('.')
print('Text To Sentences:\n',sentences,'\n')

for ch in sentences:
    # Check if 'dream' exists in the sentence (case-insensitive)
    if 'dream' in ch.lower():
        # Append to file instead of overwriting
        with open("dream_sentences.txt", "a") as f:
            f.write(ch.strip() + ".\n")  # write one sentence per line

# After loop, read back and print the results
with open("dream_sentences.txt", "r") as f:
    new_s = f.readlines()
    print('Sentences containing Dream: \n',new_s)
`````

### Activity 12: Reversing Clothing Stock List

Create a program that reads the `clothing_stock.txt` file and stores all the clothing items in a list. Then reverse the order of the items and write the reversed list into a new file called `reversed_clothing_stock.txt`.

**Steps:**
1. Read all clothing items into a list.
2. Reverse the list.
3. Write the reversed list into a new file.
<hr>

`````python
my_list = []
reversed_list = my_list[::-1]

# Open the file and read lines
with open("clothing_stock.txt", "r") as f:
    read = f.readlines()
    print(read)  # show all lines

    # Loop through each line from the file
    for line in read:
        my_list.append(line.strip())  # remove extra spaces or \n
        print(my_list)
        
reversed_list = my_list[::-1]
print('The reversed list:\n', reversed_list)

with open("reversed_clothing_stock.txt", "w") as f:
    for item in reversed_list:
        f.write(item + "\n")
`````

### Activity 21: Personal Finance App

You are creating a personal finance app that allows users to categorize their expenses and track spending by category (e.g., food, entertainment, travel). Users can add expenses, remove expenses, and see a summary of their spending. What data structure would you choose to store and organize the expenses, and why?

**Example data:**

```python
{"Food": [25, 40, 15], "Entertainment": [50, 20], "Travel": [300, 150]}
```
Steps:

1. Choose a data structure that lets users easily add and track expenses by category.
2. Write a program that allows users to input expenses, remove them, and view their total spending by category.
3. Implement a function to display a summary of all spending.
<hr>

`````python
#Name: Godwin Shambuyi Kadima
# MSIS: M01087823
`````

`````python
# Dictionary to store expenses
finance_app = {
    "Food": [25, 40, 15],
    "Entertainment": [50, 20],
    "Travel": [300, 150]
}

# Ask the user what to do
action = input("Do you want to add or remove an expense? (add/remove): ").lower()
category = input("Enter category: ")

if action == "add":
    amount = float(input("Enter expense amount: "))
    if category in finance_app:
        finance_app[category].append(amount)
    else:
        finance_app[category] = [amount]

elif action == "remove":
    amount = float(input("Enter amount to remove: "))
    if category in finance_app and amount in finance_app[category]:
        finance_app[category].remove(amount)
    else:
        print("That expense or category doesn't exist.")

# Show summary
print("\nSpending summary:")
for category, expenses in finance_app.items():
    print(f"{category}: total = ${sum(expenses)}")
`````
